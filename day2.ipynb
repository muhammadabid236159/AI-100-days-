PROJECT PATIENT MANAGMENT SYSTEM

```python
print("WELCOME TO CLINIC")

patients = []
patient_id = 1

doctors = [
    "Dr. Ahmed (Cardiologist)",
    "Dr. Sara (Dermatologist)",
    "Dr. Ali (General Physician)"
]

while True:
    print("--- MENU ---")
    print("1. Register Patient")
    print("2. View Patients")
    print("3. Appointment Booking")
    print("4. Exit")

    choice = input("Enter choice: ")

    # 1. REGISTER
    if choice == "1":
        name = input("Enter Name: ")
        age = input("Enter Age: ")
        disease = input("Enter Disease: ")

        if name == "" or age == "" or disease == "":
            print(" Please fill all fields")
            continue

        patient = {
            "id": patient_id,
            "name": name,
            "age": age,
            "disease": disease
        }

        patients.append(patient)
        print(f"Patient Registered! ID = {patient_id}")

        patient_id += 1

    # 2. VIEW PATIENTS
    elif choice == "2":
        if len(patients) == 0:
            print("⚠ No patients registered yet")
        else:
            print("--- PATIENT LIST ---")
            for p in patients:
                print(f"ID: {p['id']}, Name: {p['name']}, Age: {p['age']}, Disease: {p['disease']}")

    # 3. APPOINTMENT BOOKING
    elif choice == "3":
        print("\n--- DOCTORS LIST ---")

        i = 1
        for doctor in doctors:
            print(f"{i}. {doctor}")
            i += 1

        doc_choice = int(input("Select doctor (1-3): "))

        if doc_choice >= 1 and doc_choice <= len(doctors):
            print(f"Appointment booked with {doctors[doc_choice - 1]}")
        else:
            print("Invalid doctor selection")

    # 4. EXIT
    elif choice == "4":
        print("Exiting system...")
        break

    else:
        print("Invalid choice")
```
