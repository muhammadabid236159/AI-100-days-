```python
class Products:
    count = 0

    def __init__(self, name, price):
        self.name = name
        self.price = price
        Products.count += 1

    def getinformation(self):
        print(f"The price of {self.name} is {self.price}")

    @classmethod
    def trackobjects(cls):
        print(f"Total products in store: {cls.count}")

    @staticmethod
    def discount_calculate(price, discount):
        final_price = price - (price * discount / 100)
        print(f"Price after {discount}% discount is {final_price}")


p1 = Products("iPhone", 10000)
p1.getinformation()

p2 = Products("Samsung", 100000)
p2.getinformation()

p3 = Products("Infinix", 20000)
p3.getinformation()

Products.trackobjects()

p3.discount_calculate(20000, 10)
```

Scenario 1: Student Management System 

```python
class StudentManagementSystem:
    student = 0

    def __init__(self, name, rollnumber, marks):
        self.name = name
        self.rollnumber = rollnumber
        self.__marks = marks
        StudentManagementSystem.student += 1

    def get_information(self):
        return (
            f"Student Name: {self.name}\n"
            f"Roll Number: {self.rollnumber}\n"
            f"Marks: {self.__marks}\n"
            f"Total Students: {StudentManagementSystem.student}"
        )

    def setmarks(self, newmarks):
        if 0 <= newmarks <= 100:
            self.__marks = newmarks
        else:
            print("Invalid Marks")


s1 = StudentManagementSystem("Abid", 236159, 90)
print(s1.get_information())
s1.setmarks(87)
print(s1.get_information())


s2= StudentManagementSystem("abbas",236242,86)
print(s2.get_information())
```

Inheritance (OOP) Theory

Inheritance Object-Oriented Programming (OOP) ka ek concept hai jisme ek class doosri class ki properties (variables) aur methods (functions) ko inherit (hasil) karti hai. Is se code reuse hota hai aur programming zyada organized ho jati hai.

Definition

Inheritance ek process hai jisme ek child (derived) class, parent (base) class ke attributes aur methods ko use karti hai aur zarurat par unmein naye features bhi add kar sakti hai.

```python
class employee:
    start_time="10pm"
    end_time="6pm"
    def changetime(self,newchangetime):
        self.end_time=newchangetime

class teacher(employee):
    def __init__(self,subject):
        self.subject=subject
        super().__init__()
    

class adminstaff(employee):
    def __init__(self,role):
        self.role=role
        super().__init__()

t1=teacher("python")
admin=adminstaff("admin")

print("========== Teacher Information ==========")
print("Teacher Duty Start Time:", t1.start_time)
print("Teacher Duty End Time:", t1.end_time)
print("Teacher Subject:", t1.subject)

t1.changetime("1PM")
print("Teacher Updated Duty End Time:", t1.end_time)

print("\n========== Admin Staff Information ==========")
print("Admin Duty Start Time:", admin.start_time)
print("Admin Duty End Time:", admin.end_time)
print("Admin Staff Role:", admin.role)

admin.changetime("10PM")
print("Admin Updated Duty End Time:", admin.end_time)
```

Scenario 1: Hospital Management System ⭐⭐⭐ (Intermediate)

```python
class Employee:
    def __init__(self, name, employeeid, workinghours, shifthours):
        self.name = name
        self.employeeid = employeeid
        self.workinghours = workinghours
        self.shifthours = shifthours

    def changeshifttime(self, newshifttime):
        self.shifthours = newshifttime


class Doctor(Employee):
    def __init__(self, name, employeeid, workinghours, shifthours, specialization, patientsperday):
        super().__init__(name, employeeid, workinghours, shifthours)
        self.specialization = specialization
        self.patientsperday = patientsperday

    def showinformation(self):
        print("========== Doctor Information ==========")
        print("Doctor Name:", self.name)
        print("Employee ID:", self.employeeid)
        print("Working Hours:", self.workinghours)
        print("Shift Hours:", self.shifthours)
        print("Specialization:", self.specialization)
        print("Patients Per Day:", self.patientsperday)


class Nurse(Employee):
    def __init__(self, name, employeeid, workinghours, shifthours, wardnumber, shift):
        super().__init__(name, employeeid, workinghours, shifthours)
        self.wardnumber = wardnumber
        self.shift = shift

    def shownurseinfo(self):
        print("========== Nurse Information ==========")
        print("Nurse Name:", self.name)
        print("Employee ID:", self.employeeid)
        print("Working Hours:", self.workinghours)
        print("Shift Hours:", self.shifthours)
        print("Ward Number:", self.wardnumber)
        print("Shift:", self.shift)



# Objects


doctor1 = Doctor(
    "Ali",
    101,
    "9AM - 5PM",
    "Morning",
    "Cardiologist",
    20
)

nurse1 = Nurse(
    "Ayesha",
    201,
    "9AM - 5PM",
    "Morning",
    4,
    "Night"
)



doctor1.showinformation()

print("\nDoctor Shift Updated")
doctor1.changeshifttime("Evening")
print("New Shift:", doctor1.shifthours)



nurse1.shownurseinfo()

print("\nNurse Shift Updated")
nurse1.changeshifttime("Morning")
print("New Shift:", nurse1.shifthours)
```
