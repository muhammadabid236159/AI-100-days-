```python
class Car:
    def __init__(self, name, brand):
        self.name = name
        self.brand = brand

    def isactive(self, isactive):
        self.active = isactive

    def show(self):
        print(f"Car Name: {self.name}")
        print(f"Brand: {self.brand}")
        print(f"Active: {self.active}")


c1 = Car("Prius", "Toyota")

c1.isactive(True)   # pehle value save ki
c1.show()           # phir show mein use ki
```

```python
class CompanyLaptop:
    storage = "SSD 256"

    def __init__(self, ram, brand):
        self.ram = ram
        self.brand = brand

    def show(self):
        print(self.brand, self.ram)

    @classmethod
    def show_storage(cls):
        print(cls.storage)
    
    @staticmethod
    def calculate_price(price, quantity):
        final_price = price * quantity
        discount = final_price * 10 / 100
        payable = final_price - discount
        return final_price, discount, payable


laptop1 = CompanyLaptop(256, "Dell")

laptop1.show()
laptop1.show_storage()
finalprice=laptop1.calculate_price(10000,5)
print(finalprice)
```
