Abstraction (OOP) Theory

Abstraction Object-Oriented Programming (OOP) ka ek important concept hai. Is ka matlab hai sirf zaroori (essential) information ko user ko dikhana aur implementation (internal details) ko hide karna.

Simple words mein:

"What to do" ko dikhana aur "How to do" ko chhupana hi Abstraction kehlata hai.

Real-Life Example

Socho aap ATM Machine use karte hain.

Aap sirf PIN enter karte hain.
Amount enter karte hain.
Cash receive karte hain.

Lekin ATM ke andar transaction kaise process hoti hai, bank server se communication kaise hoti hai, ye sab details aap se hidden hoti hain.

```python
from abc import ABC, abstractmethod

class animal(ABC):
    @abstractmethod
    def makesound(self):
        pass

class dog(animal):
    def makesound(self):
        return "dog bark"


class cat(animal):
    def makesound(self):
        return "cat meoe"

dog = dog()
cat= cat()

res=dog.makesound()
print(res)
```

```python
from abc import ABC, abstractmethod

class Employee(ABC):

    @abstractmethod
    def calculateSalary(self, days, amount):
        pass


class FullTimeEmployee(Employee):

    def calculateSalary(self, days, amount):
        fulltime_salary = days * amount
        return fulltime_salary


class PartTimeEmployee(Employee):

    def calculateSalary(self, days, amount):
        parttime_salary = days * (amount / 2)
        return parttime_salary


# Objects
full = FullTimeEmployee()
part = PartTimeEmployee()

# Method Calls
print("Full Time Employee Salary:", full.calculateSalary(30, 2000))
print("Part Time Employee Salary:", part.calculateSalary(30, 2000))
```

Polymorphism (OOP)

Polymorphism ka matlab hai:

"One interface, many forms."

Ya simple words mein:

Ek hi method ka naam ho, lekin different classes us method ko apne tareeqe se implement karein.

Poly ka matlab Many aur Morph ka matlab Forms hota hai.


```python
class Dog:
    def sound(self):
        return "Dog Barks"


class Cat:
    def sound(self):
        return "Cat Meows"


class Cow:
    def sound(self):
        return "Cow Moos"


animals = [Dog(), Cat(), Cow()]

for animal in animals:
    print(animal.sound())
```

```python
class vichle:
    def start(self):
        return "start the vihcle"
    
class bike:
    def start(self):
        return "bike has started"

class car:
    def start(self):
        return "car has started"

c1=[vichle(),bike(),car()]
for car in c1:
    print(car.start()) 
```
