opreator presedence means priority order

| Priority | Operator               | Description                  |            |
| -------- | ---------------------- | ---------------------------- | ---------- |
| 1        | `()`                   | Parentheses (sab se pehle)   |            |
| 2        | `**`                   | Exponent (power)             |            |
| 3        | `+x, -x, ~x`           | Unary plus, minus, NOT       |            |
| 4        | `*, /, //, %`          | Multiply, divide, floor, mod |            |
| 5        | `+, -`                 | Addition, subtraction        |            |
| 6        | `<<, >>`               | Bit shift                    |            |
| 7        | `&`                    | Bitwise AND                  |            |
| 8        | `^`                    | Bitwise XOR                  |            |
| 9        | `                      | `                            | Bitwise OR |
| 10       | `==, !=, >, <, >=, <=` | Comparisons                  |            |
| 11       | `not`                  | Logical NOT                  |            |
| 12       | `and`                  | Logical AND                  |            |
| 13       | `or`                   | Logical OR                   |            |
