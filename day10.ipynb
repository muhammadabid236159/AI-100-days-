## 1) Anaconda kya hai?

Anaconda ek **free distribution** hai Python our R programming languages ka. Iska matlab hai ke jab hum Anaconda install karte hain to sirf Python nahi milta, balke sath mein bohat sare **useful tools our libraries** bhi already installed milte hain, jaise:

- NumPy
- Pandas
- Matplotlib
- Jupyter Notebook
- Spyder (IDE)

Isliye Data Science our Machine Learning karne walon ke liye Anaconda best choice hai, kyun ke sab kuch ek jagah mil jata hai our har library ko alag se install nahi karna parta.

### Anaconda Install Kaise Karein (Steps)

1. Browser mein jayen: **https://www.anaconda.com/download**
2. Apna operating system select karein (Windows / Mac / Linux)
3. Installer (.exe file ya .sh file) download karein
4. Installer ko open karein our **Next → I Agree → Next → Install** karte jayen
5. Installation complete hone ke baad **Anaconda Navigator** open karein confirm karne ke liye ke sab theek install hua hai

> Tip: Installation ke waqt "Add Anaconda to my PATH" wala option bhi select kar lein, taake terminal/cmd se seedha conda commands chal sakein.

## 2) Conda kya hai?

Conda ek **package our environment manager** hai. Iska kaam hai:

- Naye packages install karna
- Packages ko update ya remove karna
- Alag alag **environments** banana (har environment mein alag Python version our alag libraries ho sakti hain)

Environment banane ka fayda ye hai ke agar ek project ke liye Python 3.9 chahiye our doosre ke liye 3.11, to hum alag alag environments bana kar dono ko separately manage kar sakte hain — is se conflicts nahi hote.

### Zaroori Conda Commands (Terminal / Anaconda Prompt mein chalayen)

```bash
# Conda ka version check karna
conda --version

# Naya environment banana
conda create --name myenv python=3.10

# Environment activate karna
conda activate myenv

# Environment deactivate karna
conda deactivate

# Sare environments ki list dekhna
conda env list

# Package install karna
conda install numpy

# Package remove karna
conda remove numpy

# Sare installed packages dekhna
conda list

# Conda ko update karna
conda update conda
```

Ye commands terminal (Mac/Linux) ya Anaconda Prompt (Windows) mein chalte hain, is Jupyter code cell mein nahi — kyun ke ye system-level commands hain.

## 3) Jupyter Notebook kya hai?

Jupyter Notebook ek **web-based tool** hai jahan hum code likh sakte hain, usko turant run kar ke result dekh sakte hain, our sath mein notes (markdown) bhi likh sakte hain. Isi wajah se ye seekhne our data explore karne ke liye bohat popular hai.

Notebook mein 2 tarah ke cells hote hain:
- **Code cell** – jahan Python code likhte hain
- **Markdown cell** – jahan text/notes likhte hain (jaise ye cell)

### Jupyter Notebook Install our Open Karna

Agar Anaconda already install hai to Jupyter bhi sath mein install ho chuka hota hai. Lekin agar sirf Jupyter install karna ho to terminal mein ye command chalayen:

```bash
pip install notebook
```

Notebook open karne ke liye terminal ya Anaconda Prompt mein likhen:

```bash
jupyter notebook
```

Ye command chalate hi browser mein Jupyter Notebook automatically open ho jayega.
