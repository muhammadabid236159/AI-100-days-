What is File I/O?

File I/O means File Input and Output.

Input (I): Reading data from a file.
Output (O): Writing data into a file.

A file is used to store data permanently. Even if you close the program, the data remains saved.

Real-Life Example

Imagine a notebook.

Reading the notebook = File Input
Writing in the notebook = File Output
Why Do We Use Files?

We use files to:

Save data permanently.
Read previously saved data.
Store records like students, employees, products, etc.
Share data between programs.
Types of Files

There are two main types of files:

1. Text File (.txt)

Stores data as normal text.

Example:

Ali
20
Swat
2. Binary File (.bin)

Stores data in binary format (0s and 1s).

It is used for:

Images
Videos
Audio
PDF files
Executable programs
Opening a File

Before reading or writing, we must open the file.

Syntax:

open("filename", "mode")

Example:

file = open("student.txt", "r")
File Modes
Mode	Meaning
"r"	Read file
"w"	Write file (creates a new file or overwrites existing data)
"a"	Append data (adds data at the end)
"x"	Create a new file (gives an error if the file already exists)
"r+"	Read and write
"w+"	Read and write (overwrites old data)
"a+"	Read and append
Reading a File

Reading means getting data from a file.

Common methods:

1. read()

Reads the entire file.

file.read()
2. readline()

Reads only one line.

file.readline()
3. readlines()

Reads all lines and returns a list.

file.readlines()
Writing to a File

Writing means saving data into a file.

Method:

file.write("Hello")
Appending Data

Append means adding new data without deleting old data.

file = open("student.txt", "a")
Closing a File

After finishing, always close the file.

file.close()

Closing a file:

Saves changes.
Frees system resources.
with Statement (Recommended)

Instead of calling close(), use with.

Example:

with open("student.txt", "r") as file:
    print(file.read())

Benefits:

Automatically closes the file.
Safer and easier to use.
Prevents resource leaks.
Advantages of File I/O
Permanent data storage.
Easy to read and update data.
Useful for databases, reports, logs, and projects.
Makes programs more practical.
Disadvantages of File I/O
Reading and writing files can be slower than using variables in memory.
Errors can occur if the file does not exist or permissions are missing.
Large files may use more memory and take longer to process.

Interview/Exam Definition

File I/O (File Input/Output) is the process of reading data from a file and writing data to a file. It allows programs to store information permanently so it can be used even after the program is closed.
